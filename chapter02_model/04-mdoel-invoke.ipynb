# 关于Invoke使用

## 1. invoke的传参
### 1.1 文本输入

```python
import os

from dotenv import load_dotenv
from langchain.chat_models import init_chat_model


# 1.读取.env配置文件中的信息。相关的环境变量以.env文件中的优先
load_dotenv(override=True)

DEEPSEEK_API_KEY=os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL=os.getenv("DEEPSEEK_BASE_URL")


#2. 模型的初始化
model=init_chat_model(
    # model="deepseek-v4-flash",
    # model_provider="deepseek",
    model="deepseek:deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
)
```

```python
#3.模型的调用
print(model.invoke("你是谁？"))
```

## 2  字典列表输入

```python
messages=[{"role":"system","content":"你是一名数学老师"},
          {"role":"user","content":"帮我解释斐波那契数列"}
          ]
print(model.invoke(messages))
```

多轮对话

```python
messages=[{"role":"system","content":"你是一名数学老师"},
          {"role":"user","content":"1+2=?"},
          {"role":"assistant","content":"3"},
          {"role":"user","content":"我刚才问了什么问题?"},
          ]
print(model.invoke(messages))
```

传递记忆

```python
conversation=[
    {"role":"system","content":"你是一名数学老师"},
    {"role":"user","content":"我叫小明"}
]


response1=model.invoke(conversation)
print(response1.content)


conversation.append({"role":"assistant","content":response1.content})
conversation.append({"role":"user","content":"我叫什么名字？"})

response2=model.invoke(conversation)
print(response2.content)
```

## 3 消息列表输入

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

conversation=[
    SystemMessage(content="你是一名数学老师"),
    HumanMessage(content="我是小明")
]

response1=model.invoke(conversation)
print(response1.content)


#添加记忆
conversation.append(AIMessage(content=response1.content))
conversation.append(HumanMessage(content="我是谁"))


response2=model.invoke(conversation)
print(response2.content)
```

## 4 invoke 返回值

```python
message=[HumanMessage(content="2+3*2=?")]
response3=model.invoke(message)
print(type(response3))
```

美化输出

# 批量调用（batch）
按照顺序

```python
messages = [
"你好，你是谁?",
"2+3*5=?",
"中国首都在哪里?",
]
responses = model.batch(messages)
for response in responses:
    print(response)
```

按照完成时间顺序

```python
responses = model.batch_as_completed(messages)
for response in responses:
    print(response)
```
