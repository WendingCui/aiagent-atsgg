###  1.1 调用deepseek官网的deepseek模型

调用大模型涉及到三个重要参数：base_url,model_name,api_key

举例1：将参数信息存储在.env配置文件中


```python
import os

from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langchain_classic.callbacks.tracers import langchain
from langchain_deepseek import ChatDeepSeek
from openai import responses

# 1.读取.env配置文件中的信息。相关的环境变量以.env文件中的优先
load_dotenv(override=True)
DEEPSEEK_API_KEY=os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL=os.getenv("DEEPSEEK_BASE_URL")


#2. 模型的初始化
agent=ChatDeepSeek(
    model="deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    api_base=DEEPSEEK_BASE_URL,
)


#3.模型的调用
response=agent.invoke("你是谁？")
print(response)
```

# 4. 使用langchain1.x的统一方式

### 使用init_chat_model()

### 举例1：调用deepseek官网的DEEPSEEK模型

```python
import os

from dotenv import load_dotenv
from langchain.chat_models import init_chat_model


# 1.读取.env配置文件中的信息。相关的环境变量以.env文件中的优先
load_dotenv(override=True)

DEEPSEEK_API_KEY=os.getenv("DEEPSEEK_API_KEY")
DEEPSEEK_BASE_URL=os.getenv("DEEPSEEK_BASE_URL")


#2. 模型的初始化
model=init_chat_model(
    # model="deepseek-v4-flash",
    # model_provider="deepseek",
    model="deepseek:deepseek-v4-flash",
    api_key=DEEPSEEK_API_KEY,
    base_url=DEEPSEEK_BASE_URL,
)
 

#3.模型的调用
print(model.invoke("你是谁？"))
```
